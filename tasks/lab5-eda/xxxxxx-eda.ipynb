# EDA: Pojazdy Elektryczne

Do zajęć wykorzystano zbiór danych postępny pod adresem: 

https://www.kaggle.com/datasets/urvishahir/electric-vehicle-specifications-dataset-2025/data?select=electric_vehicles_spec_2025.csv.csv


W ramach zadania domowego, proszę o załączenie pojedynczego pliku (skryptu lub notatnika .ipynb) napisanego w języku Python, zawierającego wyniki Eksploracyjnej Analizy Danych zbioru danych o nazwie `data.csv`.

W pliku proszę opisać i wskazać poniższe elementy:
1. Informacje ogólne o zbiorze danych (liczba wierszy i kolumn, typy danych w poszczególnych kolumnach, ogólne statystyki opisowe każdej kolumny opisowej). 
2. Wykresy, w tym:  
    (a) wykres pokazujący liczbę samochodów dla 10 producentów o najwyższej liczbie modeli;  
    (b) histogram zasięgu dla różnych typów przekładni;  
    (c) moc ładowania w funkcji pojemności baterii.  
3. Odpowiedzi na pytania wraz z kodem źródłowym:  
    (a) która marka i model ma największy zasięg, moc i pojemność baterii?  
    (b) poroszę podać listę 10 pojazdów z największym przekrojem czołowym,  
    (c) proszę pokazać macierz korelacji cech z wyświetleniem jej jako wykresu typu "heatmap".
